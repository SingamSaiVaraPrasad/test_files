```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
import sklearn.datasets
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
d=sklearn.datasets.load_breast_cancer()
```

```python
df=pd.DataFrame(d.data,columns=d.feature_names)
df['labels']=d.target
df['labels'].isnull
x=df.drop(columns='labels',axis=1)
y=df['labels']
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

```python
svc=SVC()
hparameters={
    'kernel':['linear','rbf','poly','sigmoid'],
    'C':[1,5,10,20]
}
# we need to find the combination of above parameters of kernel and 'c' that gives the optimal output
# randomized Search CV is used when there are lot of combinations of HypParam are possible so we randomly select one combo
# cv stands for cross validation
classifier=GridSearchCV(svc,hparameters,cv=5)
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
#  load all this results in to pandas df
results=pd.DataFrame(classifier.cv_results_)
results
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```
